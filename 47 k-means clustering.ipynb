# load the Data

## Visualize

```python
plt.figure(figsize=(16,9))
plt.title('Unclustered Data')
plt.grid()
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.scatter(x.iloc[:,0], x.iloc[:, 1], color = 'b', marker='o')
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(random_state=0, n_clusters= 7)
```

```python
km.fit(x)
```

```python
# sum squared error
km.inertia_
```

```python
km.labels_
```

```python
sse=[]
for k in range(1,16):
    km = KMeans(random_state=0, n_clusters= k, n_init=10)
    km.fit(x)
    sse.append(km.inertia_)
```

```python
plt.title('Elbow Method')
plt.grid()
plt.xlabel('Value of k')
plt.ylabel('SSE')
plt.xticks(range(1,16))
plt.plot(range(1,16) ,sse, color='red', marker='o', mfc='green' )
```

# silhouette score

```python
from sklearn.metrics import silhouette_score
```

```python
silh = []
for k in range(2,16):
    km = KMeans(random_state=0, n_clusters= k, n_init=10)
    km.fit(x)
    score = silhouette_score(x, km.labels_)
    silh.append(km.inertia_)
```

```python
plt.title('Silhouette method')
plt.grid()
plt.xlabel('Value of k')
plt.ylabel('Sihouette score')
plt.xticks(range(2,16))
plt.bar(range(2,16) ,silh, color='red' )
```

# Create thr cluster

```python
km = KMeans(random_state=0, n_clusters= 5, n_init=10)
```

```python
y= km.fit_predict(x)
```

```python
y
```

```python
km.labels_
```

```python
km.inertia_
```

```python
cents = km.cluster_centers_
cents
```

### Visualize the clustered data

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)

plt.title('Unclustred data')
plt.grid()
plt.xlabel('annual Income')
plt.ylabel('Spending score')
plt.scatter(x.iloc[:,0], x.iloc[:,1], color= 'b', marker='o')


plt.subplot(1,2,2)
plt.title('clustred data')
plt.grid()
plt.xlabel('annual Income')
plt.ylabel('Spending score')

plt.scatter(x.iloc[:,0], x.iloc[:,1],  marker='o' , c=y, )
plt.scatter(cents[:,0], cents[:,1], color = 'cyan' , s=150 )
```
