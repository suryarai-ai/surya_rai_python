```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1
```

```python
image = cv2.imread('images/sachin.jpg')
```

```python
# get the image dimension
height, width = image.shape[:2]

# Define the translation matrix
T = np.float32([[1, 0 , 400 ],[0 , 1 , -150 ]])

# we use wrapAffine to translate the image using the matrix, T ///apply the translation
img_translation = cv2.warpAffine(image, T, (width+ 400, height + 150))
```

```python
T
```

```python
cv2.imshow('Original image', image)
cv2.imshow('translated image', img_translation)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread('images/sachin.jpg')

# rotate the image 60 deg COUNTER-CLOCK WISE:


# //////Define the centre point and rotate angle/////////

center = (image.shape[1] // 2, image.shape[0] // 2)
angle = 60
#create rotation matrix
rotated_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)

#apply the rotation using warpAffine
rotated_image = cv2.warpAffine(image, rotated_matrix,(image.shape[1], image.shape[0]))






# SHOW THWE ROTATED IMAGE
cv2.imshow('Rotated image', rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('images/sachin.jpg') 

# Resize the image to a fixed width and height
resize_image = cv2.resize(img, None, fx=0.75, fy=0.75 )  #(width, height)

# SHOW THE RESIZESD IMAGE
cv2.imshow('Original image', image)
cv2.imshow('RESIZED image', resize_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#4 FLIPPING
```

```python
# Load the image
image = cv2.imread('images/sachin.jpg')

#flip the image horizontally
flipped_horizontal = cv2.flip(image,0)

#flip the image vertically
flipped_vertical = cv2.flip(image,1)

#flip the image vdiagonally
flipped_diagonal = cv2.flip(image,-1)

# SHOW THWE flipped IMAGE
cv2.imshow('original image', image)
cv2.imshow('flipped image1', flipped_horizontal)
cv2.imshow('flipped image2',flipped_vertical)
cv2.imshow('flipped image3',flipped_diagonal)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
