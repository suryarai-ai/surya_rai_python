# kernel

```python
import cv2
import numpy as np

# load the image
image = cv2.imread('images/lenna.png')

# define a 3*3 averaging kernel
kernel = np.ones((3, 3), np.float32) /9

#apply the kernel to smooth the image
smooth_image = cv2.filter2D(image, -1, kernel)
#show the original and  to smooth th eimageed image
cv2.imshow('orignal img', image)

cv2.imshow('smoothed img', smooth_image)

cv2.waitKey(0)
cv2.destroyAllWindows()


```

# gaussian blur

```python
import cv2
import matplotlib.pyplot as plt
gray = cv2.imread('images/lenna.png')
kernel_size = 5 #adjust as needed
kernel = cv2.getGaussianKernel(kernel_size, 1.25)
blurred = cv2.filter2D(gray, -1, kernel) #apply filter
```

```python
kernel
```

```python

plt.figure(figsize=(10,9))
plt.subplot(121), plt.imshow(gray, camp='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(blurred, camp='gray'), plt.title('Gaussina blur')
#plt.show()
```
