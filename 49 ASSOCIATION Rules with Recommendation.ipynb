### Create the Dataset

```python
dataset = [
    ['Apple','Beer','Rice', 'Chicken'],
    ['Apple','Beer','Rice',],
    ['Apple','Beer'],
     ['Apple','Pear'],
    ['Milk','Beer','Rice', 'Chicken'],
    ['Milk','Beer','Rice',],
    ['Milk','Beer'],
    ['Milk','Pear']
]
```

```python
dataset
```

# Dataset

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
encoded = te.fit_transform(dataset)
```

```python
encoded 
```

```python
te.columns_
```

```python
df= pd.DataFrame(encoded, columns=te.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# 1 fIND frequent items
freq_itemset = apriori(df, min_support=0.2 , use_colnames= True )
```

```python
freq_itemset
```

```python
rules = association_rules(freq_itemset, metric='confidence', min_threshold=0.5)
```

```python
rules
```

```python
rules = rules[['antecedents', 'consequents', 'support', 'confidence']]
```

```python
rules
```

### Sort the rules

```python
rules.sort_values(by=['confidence','support'], ascending=False)
```

#### Filter the Rule

```python
nrule = rules[rules['confidence'] > 0.5]
```

```python
nrule.shape
```

```python
nrule[nrule['antecedents']=={'Rice'}]
```

```python
nrule[nrule['antecedents']=={'Rice'}]['consequents']
```

```python
nrule[nrule['antecedents']=={'Apple', 'Beer'}]['consequents']
```

# Export the rules

```python
nrule.to_csv('rules.csv', index=False)
```
