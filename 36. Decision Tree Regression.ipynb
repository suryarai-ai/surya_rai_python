```python
# import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load The Data

```python
df = pd.read_csv('datasets/petrol_consumption.csv')
```

```python
df
```

```python
df.shape
```

# Separate The input and output variable

```python
x = df.drop('Petrol_Consumption', axis=1)
y = df['Petrol_Consumption']
```

```python
x.corrwith(y)
```

```python
# none of above is more than 80% hence cant solve this by using any linear algo
#hence we will go for nonlinear algo
```

```python

sns.kdeplot(y)
```

```python
x.describe()
```

# cross validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=0)
```

```python
x_train.shape, x_test.shape
```

# Build the Model

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor(random_state=0)
```

```python
regressor.fit(x_train, y_train)
```

## Performance Evaluation

```python
y_pred = regressor.predict(x_test)
```

```python
result = pd.DataFrame({'Actual':y_test, 'Predicted': y_pred})
result
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import mean_squared_error
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
(1-mean_absolute_percentage_error(y_test, y_pred))*100
```

```python
# 87% that means we can go for linear regression
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(regressor);
```

```python
plt.figure(figsize=(12,9))
plot_tree(regressor, fontsize=6, feature_names=x.columns,
         filled= True, precision=2, class_names=['O','F'],
         rounded=True);
```

```python
plt.figure(figsize=(12,9))
plt.title('Feature Importances')
plt.xlabel('Features')
plt.ylabel('Importance')


plt.bar(x.columns, regressor.feature_importances_, color='magenta')
```
