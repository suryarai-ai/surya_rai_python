```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/lenna.png', 0)
```

```python
new_img = img > 127
```

```python
new_img
```

```python
plt.figure(figsize=(10, 9))
plt.title('Image Histogram')
plt.imshow(new_img)
```

```python
new_img = img > 127
inv_img = img < 127
```

```python
plt.figure(figsize=(16,9))

plt.subplot(131)
plt.title('original Image')


plt.imshow(img, camp='gray')


plt.subplot(132)
plt.title('binary Image')

plt.imshow(new_img, camp='gray')

plt.subplot(133)
plt.title('Inv binary Image')

plt.imshow(inv_img, camp='gray')


#coorected code mtach with this
plt.figure(figsize=(16,9))

plt.subplot(131)
plt.title('original Image')
plt.imshow(new_img, cmap='gray')

plt.subplot(142)
plt.title('Binary Image')
plt.imshow(img, cmap='gray')

plt.subplot(143)
plt.title('Inv Binary Image')
plt.imshow(inv_img, cmap='gray')


```
