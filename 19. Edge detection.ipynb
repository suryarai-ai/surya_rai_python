```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/lenna.png', 0)
```

```python
# Sobel parameters: ksize, dx, dy, depth
sobelx =cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
#sobel for horizontal edge
sobely =cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
#sobel for verticalal edge

#combine the sobel images(optional)
combined_sobel = cv2.add(sobelx, sobely)
```

```python
final_img = np.hstack([sobelx, sobely, combined_sobel])
cv2.imshow('all images', final_img)
cv2.waitKey(0)
cv2.destroyAllWindows
```

```python
plt.figure(figsize=(16,8))

plt.subplot(1,4,1);
plt.title('original Image')
plt.imshow(img, cmap='gray')

plt.subplot(1,4,2);
plt.title('sobel x')
plt.imshow(sobelx, cmap='gray')

plt.subplot(1,4,3)
plt.title('Sobel y')
plt.imshow(sobely, cmap='gray')

plt.subplot(1,4,4)
plt.title('final')
plt.imshow(combined_sobel, cmap='gray' )
```
