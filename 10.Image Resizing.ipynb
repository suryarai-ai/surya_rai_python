# using skimage 

```python
from skimage.transform import resize
import matplotlib.pyplot as plt
```

```python
## Load the image
image = plt.imread('images/lenna.png')
```

```python
# Resize the image with bilinear interpolation
resized_image = resize(image, (1000,1200), anti_aliasing=True )  #(width, height)

#resize using bicubic interpolation
resized_image = resize(image, (1300,1300), anti_aliasing=True )

# SHOW THWE ROTATED IMAGE
plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(resized_image)
```
