```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/lenna.png', 0)
```

```python
#apply canny edge detection
edges = cv2.Canny(img, 50, 150) #adjust threshold if neede

#dispaly the result
cv2.imshow('orignal img', img)

cv2.imshow('canny edge detection ', edges)

cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
import cv2

#load the image
img = cv2.imread('images/cat.jpg', 0) #load as grayscale

#apply canny edge detection
edges = cv2.Canny(img, 50, 150) #adjust threshold if neede

#dispaly the result
plt.figure(figsize=(16,8))

plt.subplot(1,2,1);
plt.title('original Image')
plt.imshow(img)

plt.subplot(1,2,2);
plt.title('canny Image detected')
plt.imshow(edges)
```

# contours detection

```python
import cv2

#load the image
img = cv2.imread('images/circle.png', 0) #load as grayscale

#create a binary image where pixels above 127 are set to 255(white),and others to 0(black)
thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)[1] 



contours,hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

cv2.drawContours(img, contours, -1, (0,255,0), 2)#draw all contours

#dispaly the result
plt.figure(figsize=(16,8))

plt.subplot(1,2,1);
plt.title('original Image')
plt.imshow(img)

plt.subplot(1,2,2);
plt.title('contour')
plt.imshow(thresh)
```

```python
import cv2

#load the image
img = cv2.imread('images/circle.png') #load as grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python

#create a binary image where pixels above 127 are set to 255(white),and others to 0(black)
thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)[1] 



contours, hierarchy = cv2.findContours(gray, thresh)

cv2.drawContours(gray, contours, -1, (0,255,0), 2)#draw all contours

#dispaly the result
plt.figure(figsize=(16,8))

plt.subplot(1,2,1);
plt.title('original Image')
plt.imshow(img)

plt.subplot(1,2,2);
plt.title('contour')
plt.imshow(thresh)
```
