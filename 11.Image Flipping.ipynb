```python
import cv2

# Load the image
image = cv2.imread('images/lenna.png')

#flip the image horizontally
flipped_horizontal = cv2.flip(image,0)

#flip the image vertically
flipped_vertical = cv2.flip(image,1)

#flip the image vdiagonally
flipped_diagonal = cv2.flip(image,-1)

# SHOW THWE flipped IMAGE
cv2.imshow('original image', image)
cv2.imshow('flipped image1', flipped_horizontal)
cv2.imshow('flipped image2',flipped_vertical)
cv2.imshow('flipped image3',flipped_diagonal)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
flip_hor = cv2.cvtColor (flipped_horizontal, cv2.COLOR_BGR2RGB)
flip_ver = cv2.cvtColor (flipped_vertical, cv2.COLOR_BGR2RGB)
flip_dia = cv2.cvtColor (flipped_diagonal, cv2.COLOR_BGR2RGB)
img_org = cv2.cvtColor (image, cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,9))

plt.subplot(1,4,1)
plt.title('orignal img')
plt.imshow(img_org)

plt.subplot(1,4,2)
plt.title('horizontal flip')
plt.imshow(flip_hor)

plt.subplot(1,4,3)
plt.title('Vertical flip')
plt.imshow(flip_ver)

plt.subplot(1,4,4)
plt.title('dIAGONAL flip')
plt.imshow(flip_dia)




```
